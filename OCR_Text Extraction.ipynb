```python
# First install required library
!pip install easyocr

import easyocr

def extract_text_with_easyocr(image_path):
    # Create reader object (auto-downloads models on first run)
    reader = easyocr.Reader(['en'])  # For English

    # Read text from image
    results = reader.readtext(image_path)

    # Combine all detected texts
    text = '\n'.join([result[1] for result in results])

    return text

# Usage
image_path = '/content/download.png'
extracted_text = extract_text_with_easyocr(image_path)

print("Extracted Text:")
print(extracted_text)
```

```python
# Step 1: Install Dependencies
!pip install easyocr pandas opencv-python-headless
!apt-get install -y poppler-utils

# Step 2: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 3: Import Libraries
import os
import cv2
import easyocr
import pandas as pd
import re
import numpy as np

# Initialize OCR Reader
reader = easyocr.Reader(['en'])

# Configure Paths
folder_path = '/content/drive/MyDrive/Images'  # Apna folder path dalen
output_csv = '/content/drive/MyDrive/extracted_data.csv'
''
# Preprocessing Functions
def preprocess_image(img):
    # Resize
    img = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
    # Invert colors
    img = cv2.bitwise_not(img)
    # Convert to grayscale
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

# Data Extraction Logic
def process_image(img_path):
    # Read and preprocess image
    img = cv2.imread(img_path)
    if img is None:
        return {}
    img = preprocess_image(img)

    # OCR with coordinates
    results = reader.readtext(img, detail=1)

    data = {"First Name":"", "Father Name":"", "Identity Number":""}
    labels = {"name": [], "father": []}
    cnic_pattern = re.compile(r'\d{5}-\d{7}-\d')

    # First Pass: Detect Labels and CNIC
    for (bbox, text, _) in results:
        y_center = (bbox[0][1] + bbox[2][1])/2
        text_clean = text.lower().replace(":", "")

        if 'father' in text_clean and 'name' in text_clean:
            labels['father'].append((y_center, text))
        elif 'name' in text_clean:
            labels['name'].append((y_center, text))
        elif cnic_pattern.search(text):
            data['Identity Number'] = cnic_pattern.search(text).group()

    # Second Pass: Find Values
    for y, text in labels['name']:
        candidates = []
        for (bbox, val, _) in results:
            if bbox[0][1] > y + 10:  # At least 10 pixels below label
                clean_val = val.split(":")[-1].strip()
                if clean_val and len(clean_val) > 1:
                    candidates.append(clean_val)
        if candidates:
            data['First Name'] = candidates[0]

    for y, text in labels['father']:
        candidates = []
        for (bbox, val, _) in results:
            if bbox[0][1] > y + 10:
                clean_val = val.split(":")[-1].strip()
                if clean_val and len(clean_val) > 1:
                    candidates.append(clean_val)
        if candidates:
            data['Father Name'] = candidates[0]

    return data

# Process All Images
all_data = []
for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(folder_path, filename)
        print(f"Processing {filename}...")
        try:
            result = process_image(img_path)
            all_data.append(result)
        except Exception as e:
            print(f"Error processing {filename}: {str(e)}")

# Save to CSV
df = pd.DataFrame(all_data)
df.to_csv(output_csv, index=False)
print("\nProcessing Complete! CSV Saved to Google Drive.")
print("Final Results:")
print(df)
```
